## Grouping Data with pandas

**1-Split the data into groups based on some criteria-group data based on product_name column**

```python
df.groupby('product_name')
```

## Aggregating data with agg()

**calculate the mean of the "order_number" column grouped by the "department_id" column**

```python
df.groupby('department_id').agg({'order_number': ['mean']})
```

**same as above, but using .mean function**

```python
df.groupby('department_id')['order_number'].mean()
```

**multiple aggregations**

```python
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

## Aggregating Data with transform()

**Split the data into groups based on the “user_id” column; Apply the transform() function on the “order_number” column to generate the maximum orders for each user; Create a new column, “max_order,” into which you’ll place the results of your aggregation.**

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby('user_id')['order_number'].transform(np.max)
```

**check results**

```python
ords_prods_merge.head(15)
```

## Deriving Columns with loc()

**flagging customers depending on max_order**

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

**check the frequency of each loyalty flag**

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```
